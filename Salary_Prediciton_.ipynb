```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#defining data frame and column names since column names are absent in the raw data
data = pd.read_csv("/content/adult.data.csv", header=None, sep=',\s', na_values=["?"])
data.columns = ['Age','Work Class', 'Final Weight', 'Education', 'Experience','Marital Status', 'Occupation',"Relationship", 'Race','Gender','Capital Gain','Capital Loss','Work hours/Week','Country of Origin','Income']
```

```python
data.head()
```

```python
#converting dummy values from categorical data
workclass_dummy = pd.get_dummies(data[["Work Class"]],drop_first = True)
data = pd.concat([data,workclass_dummy],axis=1)

education_dummy = pd.get_dummies(data[["Education"]],drop_first = True)
data = pd.concat([data,education_dummy],axis=1)

maritalstatus_dummy = pd.get_dummies(data[["Marital Status"]],drop_first = True)
data = pd.concat([data,maritalstatus_dummy],axis=1)

occupation_dummy =pd.get_dummies(data[["Occupation"]],drop_first = True)
data = pd.concat([data,occupation_dummy],axis=1) 

relationship_dummy =pd.get_dummies(data[["Relationship"]],drop_first = True)
data = pd.concat([data,relationship_dummy],axis=1)

race_dummy = pd.get_dummies(data[["Race"]],drop_first = True)
data = pd.concat([data,race_dummy],axis=1) 

country_of_origin_dummy = pd.get_dummies(data[["Country of Origin"]],drop_first = True) 
data = pd.concat([data,country_of_origin_dummy],axis=1)
```

```python
data.head()
```

```python
# Converting Male and female to 1 and 0
def binary(x):
  if x == "Male":
    return 1
  else:
    return(0)  
```

```python
data["Gender"]= data["Gender"].apply(binary) 
```

```python
data.head()
```

```python
#dropping all categorical values
data.drop(['Work Class','Education','Marital Status','Occupation','Relationship','Race','Country of Origin'],axis=1,inplace=True)
```

```python
data.head()
```

```python
#moving income column to the last column
income = data.pop("Income")
```

```python
data.insert(97,"Income",income)
```

```python
data.head()
```

```python
#droping all the NaN values which were converted from '?'
data=data. dropna()
```

```python
data = data.fillna(lambda x: x.median())
```

```python
#Defining x and y values for KNN 
x = data.iloc[:,:-1].values
y = data.iloc[:,7].values
```

```python
data.head
```

```python
#importing models, defining test,train dataset
X_test,X_train,y_test,y_train = train_test_split(x,y,test_size=0.20)
```

```python
#Normalization
scaler = StandardScaler ()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Setting KNN with our dataset and set K =15
classifier = KNeighborsClassifier(n_neighbors=15)
classifier.fit(X_train,y_train)
```

```python
#predictions
y_pred = classifier.predict(X_test)
```

```python
#print f1 score and confusion matrx to determine accuracy
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```
